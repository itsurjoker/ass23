```python
#Q2.

#A scatter plot is a graphical representation used to display the relationship between two variables. It is particularly useful for visualizing how one variable changes in relation to another. Each point on the scatter plot corresponds to a pair of values, one for each of the variables being compared.

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.scatter(x, y, c='blue', marker='o', label='Data Points')

# Add a title and labels
plt.title('Scatter Plot of x vs y')
plt.xlabel('x')
plt.ylabel('y')

# Display the legend
plt.legend()

# Display the plot
plt.show()
```

```python
#Q3.

#The subplot() function is used to create multiple plots within a single figure. It allows you to arrange multiple plots in a grid or other configurations, making it easier to compare and analyze different datasets or visualizations. Each subplot is like a separate plotting area within the main figure.

import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 subplot grid
plt.figure(figsize=(10, 8))

# Plot 1
plt.subplot(2, 2, 1)
plt.plot(x1, y1, marker='o')
plt.title('Line Plot 1')

# Plot 2
plt.subplot(2, 2, 2)
plt.plot(x2, y2, marker='o')
plt.title('Line Plot 2')

# Plot 3
plt.subplot(2, 2, 3)
plt.plot(x3, y3, marker='o')
plt.title('Line Plot 3')

# Plot 4
plt.subplot(2, 2, 4)
plt.plot(x4, y4, marker='o')
plt.title('Line Plot 4')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
#Q4.

#A bar plot (also known as a bar chart or bar graph) is a graphical representation of data that uses rectangular bars to show the values of different categories. The length or height of each bar is proportional to the value it represents. Bar plots are commonly used to compare and display the relative sizes of different data points or categories.

#Bar plots are used for various purposes, including:

#   1. Comparison: Bar plots are useful for comparing the values of different categories or groups. They make it easy to visually identify which category has the highest or lowest value.

#   2. Distribution: Bar plots can show the distribution of data within different categories, helping to visualize patterns or trends.

#   3. Frequency: Bar plots can be used to display the frequency or count of data points falling into different categories.

#   4. Change Over Time: When used in a stacked or grouped format, bar plots can effectively show changes in values over time or across different conditions.


import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='blue')
plt.xlabel('Company')
plt.ylabel('Profit ($)')
plt.title('Profit Comparison by Company')
plt.show()

# Create a horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='green')
plt.xlabel('Profit ($)')
plt.ylabel('Company')
plt.title('Profit Comparison by Company (Horizontal)')
plt.show()
```

```python
#Q5.

#A box plot, also known as a box-and-whisker plot, is a graphical representation of a data set that provides a visual summary of its distribution. It displays the five-number summary of a dataset: the minimum, first quartile (25th percentile), median (second quartile or 50th percentile), third quartile (75th percentile), and maximum. The box plot consists of a rectangular box (the interquartile range or IQR) and two "whiskers" that extend from the box to indicate the variability outside the IQR.

#Box plots are commonly used to:

#   1. Visualize Distribution: Box plots provide a clear and concise way to understand the distribution of data, including the central tendency (median) and spread (IQR) of the data.

#   2. Identify Outliers: Outliers, which are values significantly different from the rest of the data, can be easily identified using box plots.

#   3. Compare Distributions: Box plots allow for easy comparison of distributions between different groups or datasets.

#   4. Detect Skewness: The asymmetry of the box plot can indicate the skewness of the data distribution.

#   5. Display Summary Statistics: The five-number summary and additional information about the data, like median, quartiles, and range, are displayed in a box plot.


import numpy as np
import matplotlib.pyplot as plt

# Generate the data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a list of data to plot
data_to_plot = [box1, box2]

# Create a box plot
plt.boxplot(data_to_plot, labels=["Box 1", "Box 2"])

# Add a title and labels
plt.title("Box Plot of Data")
plt.xlabel("Box")
plt.ylabel("Value")

# Show the plot
plt.show()
```
